#3. ÏÉùÏùº Îç∞Ïù¥ÌÑ∞Î•º ÍµêÏû¨ ÏΩîÎìú(heap.py)Î•º Ïù¥Ïö©Ìï¥ ÌûôÏúºÎ°ú Ï†ÄÏû•ÌïòÍ≥†, ÏÉùÏùºÏù¥ ÎäêÎ¶∞ ÏàúÏÑúÎ°ú 10Î™ÖÏùò ÏπúÍµ¨Î•º Ï∂úÎ†•ÌïòÎäî ÏΩîÎìúÎ•º ÏûëÏÑ±ÌïúÎã§. Ïã§Ìñâ Í≤∞Í≥ºÍ∞Ä ÏÖÄÏóê ÎÇòÌÉÄÎÇòÏïº ÌïúÎã§.

```python
import csv
import urllib.request

class Heap:
    def __init__(self):
        self.__A = []

    def insert(self, x):
        self.__A.append(x)
        self.__percolateUp(len(self.__A) - 1)

    def __percolateUp(self, i):
        parent = (i - 1) // 2
        if i > 0 and self.__A[i][0] > self.__A[parent][0]:  # Î¨∏ÏûêÏó¥ ÎπÑÍµê
            self.__A[i], self.__A[parent] = self.__A[parent], self.__A[i]
            self.__percolateUp(parent)

    def deleteMax(self):
        if not self.isEmpty():
            max = self.__A[0]
            self.__A[0] = self.__A.pop()
            self.__percolateDown(0)
            return max
        else:
            return None

    def __percolateDown(self, i):
        child = 2 * i + 1
        right = 2 * i + 2
        if child <= len(self.__A) - 1:
            if right <= len(self.__A) - 1 and self.__A[child][0] < self.__A[right][0]:
                child = right
            if self.__A[i][0] < self.__A[child][0]:
                self.__A[i], self.__A[child] = self.__A[child], self.__A[i]
                self.__percolateDown(child)

    def isEmpty(self):
        return len(self.__A) == 0

# GitHub CSV ÏùΩÍ∏∞
url = "https://raw.githubusercontent.com/o9ioakd/week6HW/main/birthday.csv"
heap = Heap()

with urllib.request.urlopen(url) as f:
    lines = [l.decode('utf-8') for l in f.readlines()]
    reader = csv.DictReader(lines)
    for row in reader:
        name = row['Ïù¥Î¶Ñ']
        birthday = row['ÏÉùÎÖÑÏõîÏùº8ÏûêÎ¶¨(Ïòà.20040101)']
        if len(birthday) == 8 and birthday.isdigit():
            heap.insert((birthday, name))

print("üéâ ÏÉùÏùºÏù¥ Îä¶ÏùÄ Ïàú Top 10 ÏπúÍµ¨:")
for i in range(10):
    if not heap.isEmpty():
        birthday, name = heap.deleteMax()
        formatted_bday = f"{birthday[:4]}ÎÖÑ {birthday[4:6]}Ïõî {birthday[6:]}Ïùº"
        print(f"{i+1}. Ïù¥Î¶Ñ: {name}, ÏÉùÏùº: {formatted_bday}")
```

```python
from bidirectNode import BidirectNode

class CircularDoublyLinkedList:
	def __init__(self):
		self.__head = BidirectNode("dummy", None, None)
		self.__head.prev = self.__head
		self.__head.next = self.__head
		self.__numItems = 0
 
	def insert(self, i:int, newItem) -> None:
		if (i >= 0 and i <= self.__numItems):
			prev = self.getNode(i - 1)
			newNode = BidirectNode(newItem, prev, prev.next)
			newNode.next.prev = newNode
			prev.next = newNode
			self.__numItems += 1
		else:
			print("index", i, ": out of bound in insert()") # ÌïÑÏöî Ïãú ÏóêÎü¨ Ï≤òÎ¶¨

	def append(self, newItem) -> None:
		prev = self.__head.prev
		newNode = BidirectNode(newItem, prev, self.__head)
		prev.next = newNode
		self.__head.prev = newNode
		self.__numItems += 1

	def pop(self, *args):
		# Í∞ÄÎ≥Ä ÌååÎùºÎØ∏ÌÑ∞. Ïù∏ÏûêÍ∞Ä ÏóÜÍ±∞ÎÇò -1Ïù¥Î©¥ ÎßàÏßÄÎßâ ÏõêÏÜåÎ°ú Ï≤òÎ¶¨ÌïòÍ∏∞ ÏúÑÌï®. ÌååÏù¥Ïç¨ Î¶¨Ïä§Ìä∏ Í∑úÏπô ÎßåÏ°±
		if self.isEmpty():
			return None
		# Ïù∏Îç±Ïä§ i Í≤∞Ï†ï
		if len(args) != 0: # pop(k)Í≥º Í∞ôÏù¥ Ïù∏ÏûêÍ∞Ä ÏûàÏúºÎ©¥ i = k Ìï†Îãπ
			i = args[0]
		if len(args) == 0 or i == -1:# pop()Ïóê Ïù∏ÏûêÍ∞Ä ÏóÜÍ±∞ÎÇò pop(-1)Ïù¥Î©¥ iÏóê Îß® ÎÅù Ïù∏Îç±Ïä§ Ìï†Îãπ
			i = self.__numItems - 1
		# iÎ≤à ÏõêÏÜå ÏÇ≠Ï†ú
		if (i >= 0 and i <= self.__numItems - 1):
			curr = self.getNode(i)
			retItem = curr.item
			curr.prev.next = curr.next
			curr.next.prev = curr.prev
			self.__numItems -= 1
			return retItem
		else:
			return None
 
	def remove(self, x):
		curr = self.__findNode(x)
		if curr != None:
			curr.prev.next = curr.next
			curr.next.prev = curr.prev
			self.__numItems -= 1
			return x
		else:
			return None

	def get(self, *args):
		# Í∞ÄÎ≥Ä ÌååÎùºÎØ∏ÌÑ∞. Ïù∏ÏûêÍ∞Ä ÏóÜÍ±∞ÎÇò -1Ïù¥Î©¥ ÎßàÏßÄÎßâ ÏõêÏÜåÎ°ú Ï≤òÎ¶¨ÌïòÍ∏∞ ÏúÑÌï®. ÌååÏù¥Ïç¨ Î¶¨Ïä§Ìä∏ Í∑úÏπô ÎßåÏ°±
		if self.isEmpty(): return None
		# Ïù∏Îç±Ïä§ i Í≤∞Ï†ï
		if len(args) != 0:   # pop(k)Í≥º Í∞ôÏù¥ Ïù∏ÏûêÍ∞Ä ÏûàÏúºÎ©¥ i = k Ìï†Îãπ
			i = args[0]
		if len(args) == 0 or i == -1::# pop()Ïóê Ïù∏ÏûêÍ∞Ä ÏóÜÍ±∞ÎÇò pop(-1)Ïù¥Î©¥ iÏóê Îß® ÎÅù Ïù∏Îç±Ïä§ Ìï†Îãπ
			i = self.__numItems - 1
		# iÎ≤à ÏõêÏÜå Î¶¨ÌÑ¥
		if (i >= 0 and i <= self.__numItems - 1):
			return self.getNode(i).item
		else:
			return None
 
	def index(self, x) -> int:
		cnt = 0
		for element in self:
			if element == x:
				return cnt
			cnt += 1
		return -12345

	def isEmpty(self) -> bool:
		return self.__numItems == 0
 
	def size(self) -> int:
		return self.__numItems

	def clear(self):
		self.__head = BidirectNode("dummy", None, None)
		self.__head.prev = self.__head
		self.__head.next = self.__head
		self.__numItems = 0
 
	def count(self, x) -> int:
		cnt = 0
		for element in self:
			if element == x:
					cnt += 1
		return cnt

	def extend(self, a): # aÎäî ÏàúÌöå Í∞ÄÎä•Ìïú Î™®Îì† Í∞ùÏ≤¥
		for element in a:
			self.append(element)

	def copy(self) -> 'CircularDoublyLinkedList':
		a = CircularDoublyLinkedList()
		for element in self:
			a.append(element)
		return a
 
	def reverse(self) -> None:
		prev = self.__head; curr = prev.next; next = curr.next
		self.__head.next = prev.prev; self.__head.prev = curr
		for i in range(self.__numItems):
			curr.next = prev; curr.prev = next
			prev = curr; curr = next; next = next.next
 
	def sort(self) -> None:
		a = []
		for element in self:
			a.append(element)
		a.sort()
		self.clear()
		for element in a:
			self.append(element)

	def __findNode(self, x) -> BidirectNode:
		curr = self.__head.next  # 0Î≤à ÎÖ∏Îìú
		while curr != self.__head:
 
			if curr.item == x:
				return curr
			else:
				curr = curr.next
		return None

	def getNode(self, i:int) -> BidirectNode:
		curr = self.__head  # ÎçîÎØ∏ Ìó§Îìú, index: -1
		for index in range(i + 1):
			curr = curr.next
		return curr

	def printList(self) -> None:
		for element in self:
			print(element, end = ' ')
		print()
 
 def __iter__(self):  # generating iterator and return
		return CircularDoublyLinkedListIterator(self)
 
class CircularDoublyLinkedListIterator:
	def __init__(self, alist):
		self.__head = alist.getNode(-1)  		# ÎçîÎØ∏ Ìó§Îìú
		self.iterPosition = self.__head.next  	# 0Î≤à ÎÖ∏Îìú
	def __next__(self):
		if self.iterPosition == self.__head: 	# ÏàúÌôò ÎÅù
			raise StopIteration
		else: # ÌòÑÏû¨ ÏõêÏÜåÎ•º Î¶¨ÌÑ¥ÌïòÎ©¥ÏÑú Îã§Ïùå ÏõêÏÜåÎ°ú Ïù¥Îèô
			item = self.iterPosition.item
			self.iterPosition = self.iterPosition.next
			return item

# ÏΩîÎìú 5-25

import csv
import urllib.request
from circularDoublyLinkedList import CircularDoublyLinkedList

# Ï°∞Ïõê Ïù¥Î¶Ñ Î¶¨Ïä§Ìä∏
my_team = [
    "ÏÜ°ÎØºÏÑú", "ÏïàÏàòÎØº", "Ïò§ÏòàÏ§Ä", "ÍπÄÏ±ÑÎØº", "ÍπÄÏòàÎπà", "Ïã†Ìù¨ÏòÅ", 
    "ÍπÄÏÑ†ÎØº", "ÍπÄÌòúÏù∏", "ÍπÄÏ£ºÌïò", "ÍπÄÎØºÏ£º", "ÏµúÍ∞ÄÏò®", "Î∞∞ÏãúÏùÄ", 
    "Í∞ïÏàòÏïÑ", "ÍπÄÏÑúÎπà"
]

# CircularDoublyLinkedList ÏÉùÏÑ±
cdll = CircularDoublyLinkedList()

# CSV ÏùΩÍ∏∞ (GitHubÏóêÏÑú)
url = "https://raw.githubusercontent.com/o9ioakd/week6HW/main/birthday.csv"
with urllib.request.urlopen(url) as f:
    lines = [l.decode('utf-8') for l in f.readlines()]
    reader = csv.DictReader(lines)
    for row in reader:
        name = row['Ïù¥Î¶Ñ']
        birthday = row['ÏÉùÎÖÑÏõîÏùº8ÏûêÎ¶¨(Ïòà.20040101)']
        if name and birthday:
            cdll.append((name, birthday))

# Ï°∞Ïõê Ïù¥Î¶Ñ + ÏÉùÏùº Ï∂úÎ†• Ìï®Ïàò
print("üë• Í∞ôÏùÄ Ï°∞ ÏπúÍµ¨Îì§ (ÏÉùÏùº Îç∞Ïù¥ÌÑ∞ ÏûàÎäî Í≤ΩÏö∞Îßå):")
for name, birthday in cdll:
    if name in my_team and len(birthday) == 8 and birthday.isdigit():
        print(f"Ïù¥Î¶Ñ: {name}, ÏÉùÏùº: {birthday[:4]}ÎÖÑ {birthday[4:6]}Ïõî {birthday[6:]}Ïùº")

```

#6. LeetCode 703.Kth Largest Element in Stream Î¨∏Ï†úÎ•º ÌëºÎã§.

```python
class MinHeap:
    def __init__(self):
        self.heap = []

    def push(self, val):
        self.heap.append(val)
        self._heapify_up(len(self.heap) - 1)

    def pop(self):
        if len(self.heap) == 0:
            return None
        if len(self.heap) == 1:
            return self.heap.pop()

        top = self.heap[0]
        # Swap first and last, then remove last
        self.heap[0] = self.heap.pop()
        self._heapify_down(0)
        return top

    def peek(self):
        return self.heap[0] if self.heap else None

    def size(self):
        return len(self.heap)

    def _heapify_up(self, index):
        while index > 0:
            parent = (index - 1) // 2
            if self.heap[index] < self.heap[parent]:
                self.heap[index], self.heap[parent] = self.heap[parent], self.heap[index]
                index = parent
            else:
                break

    def _heapify_down(self, index):
        size = len(self.heap)
        while index < size:
            smallest = index
            left = 2 * index + 1
            right = 2 * index + 2

            if left < size and self.heap[left] < self.heap[smallest]:
                smallest = left
            if right < size and self.heap[right] < self.heap[smallest]:
                smallest = right

            if smallest != index:
                self.heap[index], self.heap[smallest] = self.heap[smallest], self.heap[index]
                index = smallest
            else:
                break

class KthLargest:
    def __init__(self, k: int, nums: list[int]):
        self.k = k
        self.min_heap = MinHeap()
        for num in nums:
            self.add(num)

    def add(self, val: int) -> int:
        self.min_heap.push(val)
        if self.min_heap.size() > self.k:
            self.min_heap.pop()
        return self.min_heap.peek()

```
